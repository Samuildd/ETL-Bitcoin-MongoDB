**Checking Outlier Values Using Z-Score**

The outliers within the minute_change column are compared by the previous 60 minutes(1 hour). The first 60 rows are not going to be considered as outliers as they're considered as entry information.

In a live pipeline when a the datasets is updated and a new row is entered of bitcoin information. The close value is compared with the previous row and the change is used to calculate the last hour's average and standard deviation.

z = (x-μ)/σ <- Z-score formula
x <- value
μ <- mean
σ <- standard deviation

The z score formula is applied to the current minute_change value using the last 60 minutes mean and standard deviation.

The z_score threshold chosen is 3 which accounts for 99.7% of the data. A z_score threshold of 1 represents 68% of the data and 2 represents 95% of the data. The reason 3 was chosen due to bitcoin's volatile attributes of random increases/decreases as well as it won't over identify unncessary "outliers" and really looks at the values that fall within 0.03%. Essentially the tails of distribution.

I have developed a new column called is_outlier which takes the absolute value of z_score and runs a boolean check if it greater than 3.

Now all values that are considered outliers can be checked and verified if they really need to be removed from the dataset.

```python
bitcoin_df['last_60_mean'] = bitcoin_df['minute_change'].rolling(window=60).mean()
bitcoin_df['last_60_std'] = bitcoin_df['minute_change'].rolling(window=60).std()
```

```python
bitcoin_df['z_score'] = (bitcoin_df['minute_change'] - bitcoin_df['last_60_mean']) / (bitcoin_df['last_60_std'])
```

```python
bitcoin_df['is_outlier'] = bitcoin_df['z_score'].abs() > 3
```

```python
def change_date(x):
  if x == 0:
    return 'Monday'
  elif x == 1:
      return 'Tuesday'
  elif x == 2:
    return 'Wednesday'
  elif x == 3:
    return 'Thursday'
  elif x == 4:
    return "Friday"
  elif x == 5:
    return "Saturday"
  elif x == 6:
    return "Sunday"
  elif x == None:
    return None
#Defining function to transform from numerical day of week to string.
```

Prior to running and applying my change_date function to the bitcoin dataset, I followed the process of unit testing ensuring the code quality is up to standard.

I developed a function called testing_date_func where I run my change_date function verifying each input and output is correct. It ran through all numerical day values with the change_date function and asserted the result out. If the asserted result is different to the expected output in the change_date function an 'AssertionError' will pop up displaying where the error occurred.

If all tests are correct, the final output will be "All Unit Tests Passed".

```python
def testing_date_func():
  result = change_date(0)
  assert result == 'Monday'
  result = change_date(1)
  assert result == 'Tuesday'
  result = change_date(2)
  assert result == 'Wednesday'
  result = change_date(3)
  assert result == 'Thursday'
  result = change_date(4)
  assert result == 'Friday'
  result = change_date(5)
  assert result == 'Saturday'
  result = change_date(6)
  assert result == 'Sunday'
  result = change_date(None)
  assert result == None
  print('All Unit Tests Passed')
```

```python
testing_date_func()
#Running the testing_date_func
```
